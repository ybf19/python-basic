# 🧠 Operadores Lógicos: and, or, not

## 🎯 Objetivos
- Combinar múltiples condiciones con `and`
- Usar `or` para alternativas
- Negar condiciones con `not`

```python
# AND: todas las condiciones deben ser True
edad = 25
tiene_licencia = True
tiene_seguro = True

if edad >= 18 and tiene_licencia and tiene_seguro:
    print("✅ Puedes conducir")
else:
    print("❌ No puedes conducir")
    
    if edad < 18:
        print("- Eres menor de edad")
    if not tiene_licencia:
        print("- No tienes licencia")
    if not tiene_seguro:
        print("- No tienes seguro")
```

```python
# OR: al menos una condición debe ser True
es_fin_de_semana = False
es_vacaciones = True

if es_fin_de_semana or es_vacaciones:
    print("¡Puedes descansar! 🎉")
else:
    print("Día de trabajo 💼")
```

```python
# NOT: invierte el valor booleano
activo = False
if not activo:
    print("La cuenta está inactiva, por favor actívala")

# NOT también se puede usar con otros operadores
edad = 15
if not edad >= 18:
    print("Eres menor de edad (usando NOT)")
```

```python
# Combinando operadores
usuario = "admin"
intentos = 3
bloqueado = False

if usuario == "admin" and intentos < 5 and not bloqueado:
    print("✅ Acceso permitido al panel admin")
else:
    print("❌ Acceso denegado")
```

```python
# Tabla de verdad práctica
print("Tabla de AND:")
print(True and True)    # True
print(True and False)   # False
print(False and True)   # False
print(False and False)  # False

print("\nTabla de OR:")
print(True or True)     # True
print(True or False)    # True
print(False or True)    # True
print(False or False)   # False

print("\nTabla de NOT:")
print(not True)         # False
print(not False)        # True
```
