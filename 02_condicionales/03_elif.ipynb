# 🔀 elif: Múltiples condiciones

## 🎯 Objetivos
- Usar `elif` para múltiples caminos
- Evaluar varias condiciones en cadena

```python
# Estructura con elif
nota = 85

if nota >= 90:
    print("Calificación: A (Excelente)")
elif nota >= 80:
    print("Calificación: B (Bueno)")
elif nota >= 70:
    print("Calificación: C (Suficiente)")
elif nota >= 60:
    print("Calificación: D (Insuficiente)")
else:
    print("Calificación: F (Reprobado)")
```

```python
# Ejemplo: Clasificador de edades
edad = 25

if edad < 12:
    print("Eres un niño 👶")
elif edad < 18:
    print("Eres adolescente 🧑")
elif edad < 30:
    print("Eres joven adulto 🧑‍💼")
elif edad < 60:
    print("Eres adulto 👨")
else:
    print("Eres adulto mayor 👴")
```

```python
# Ejemplo: Días de la semana
dia = "martes"

if dia == "lunes":
    print("Comienza la semana 💪")
elif dia == "martes" or dia == "miércoles" or dia == "jueves":
    print("Mitad de semana 📚")
elif dia == "viernes":
    print("¡Por fin es viernes! 🎉")
elif dia == "sábado" or dia == "domingo":
    print("Finde semana 🎈")
else:
    print("Día no válido")
```
