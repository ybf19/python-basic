## 🎯 Objetivos
- Poner condicionales dentro de condicionales
- Manejar lógica más compleja

```python
# Condicionales anidadas
tiene_dinero = True
hace_sol = False

if tiene_dinero:
    print("Tengo dinero 💰")
    if hace_sol:
        print("Voy a la playa 🏖️")
    else:
        print("Voy al cine 🎬")
else:
    print("No tengo dinero 💸")
    print("Me quedo en casa 🏠")
```

```python
# Ejemplo: Sistema de acceso con múltiples validaciones
usuario = "admin"
contraseña = "1234"
tiene_permiso = True

if usuario == "admin":
    print("Usuario admin detectado")
    if contraseña == "1234":
        print("Contraseña correcta")
        if tiene_permiso:
            print("✅ Acceso TOTAL concedido")
            print("Bienvenido al panel de administración")
        else:
            print("❌ No tienes permisos suficientes")
    else:
        print("❌ Contraseña incorrecta")
else:
    print("❌ Usuario no reconocido")
```

```python
# Ejemplo práctico: Recomendación de vestimenta
temperatura = 28
llueve = True

if temperatura > 25:
    print("Hace calor 🌡️")
    if llueve:
        print("Recomendación: Camiseta + Paraguas ☔")
    else:
        print("Recomendación: Camiseta y shorts 🩳")
elif temperatura > 15:
    print("Temperatura agradable 🌤️")
    if llueve:
        print("Recomendación: Chaqueta ligera + Paraguas")
    else:
        print("Recomendación: Chaqueta ligera")
else:
    print("Hace frío ❄️")
    if llueve:
        print("Recomendación: Abrigo + Paraguas + Bufanda")
    else:
        print("Recomendación: Abrigo y gorro")
```
