# 🏥 PROYECTO: Calculadora de IMC + Recomendaciones

## 🎯 Objetivo del proyecto
Crear una calculadora de IMC completa que incluya:
- Cálculo del IMC
- Clasificación
- Recomendaciones personalizadas
- Historial de consultas (opcional)

```python
# Proyecto completo: Calculadora IMC
import datetime

print("=" * 50)
print("     🏥 CALCULADORA DE IMC PROFESIONAL")
print("=" * 50)

# Datos del usuario
nombre = input("\n📝 Tu nombre: ")
edad = int(input("📅 Tu edad: "))
peso = float(input("⚖️  Tu peso (kg): "))
altura = float(input("📏 Tu altura (m): "))

# Cálculo IMC
imc = peso / (altura ** 2)

# Clasificación y recomendaciones
print("\n" + "=" * 50)
print(f"📊 RESULTADOS PARA: {nombre.upper()}")
print("=" * 50)
print(f"📅 Fecha: {datetime.datetime.now().strftime('%d/%m/%Y %H:%M')}")
print(f"📊 Tu IMC es: {imc:.2f}")

if imc < 18.5:
    print("🔴 Clasificación: BAJO PESO")
    print("\n📋 Recomendaciones:")
    print("  • Aumenta tu ingesta calórica de forma saludable")
    print("  • Incluye más proteínas en tu dieta")
    print("  • Consulta a un nutricionista")
    print("  • Realiza ejercicios de fuerza")
elif imc < 25:
    print("🟢 Clasificación: PESO NORMAL")
    print("\n📋 Recomendaciones:")
    print("  • Mantén tu alimentación balanceada")
    print("  • Sigue haciendo ejercicio regularmente")
    print("  • Duerme 7-8 horas diarias")
    print("  • Bebe suficiente agua")
elif imc < 30:
    print("🟠 Clasificación: SOBREPESO")
    print("\n📋 Recomendaciones:")
    print("  • Reduce el consumo de azúcares y grasas")
    print("  • Aumenta tu actividad física")
    print("  • Camina al menos 30 minutos diarios")
    print("  • Consulta a un especialista")
else:
    print("🔴 Clasificación: OBESIDAD")
    print("\n📋 Recomendaciones:")
    print("  • ES IMPORTANTE CONSULTAR A UN MÉDICO")
    print("  • Inicia con ejercicios de bajo impacto")
    print("  • Lleva un diario de alimentos")
    print("  • Busca apoyo profesional")

# Rango saludable
peso_min = 18.5 * (altura ** 2)
peso_max = 24.9 * (altura ** 2)

print("\n📈 Tu peso saludable debería estar entre:")
print(f"   Mínimo: {peso_min:.1f} kg")
print(f"   Máximo: {peso_max:.1f} kg")
print(f"   Diferencia: {peso_max - peso_min:.1f} kg")

# Mensaje motivacional
print("\n" + "=" * 50)
print("💪 MENSAJE MOTIVACIONAL")
print("=" * 50)
mensajes = [
    "¡Cada pequeño paso cuenta hacia una vida más saludable!",
    "La constancia es más importante que la intensidad",
    "Tu salud es tu mayor riqueza",
    "Hoy es un buen día para empezar a cuidarte",
    "No se trata de ser perfecto, sino de ser mejor que ayer"
]

import random
print(f"✨ {random.choice(mensajes)}")

print("\n" + "=" * 50)
print("     🏥 ¡GRACIAS POR USAR LA CALCULADORA!")
print("=" * 50)
```
