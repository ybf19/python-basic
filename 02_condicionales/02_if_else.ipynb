# 🔀 if-else: Dos caminos

## 🎯 Objetivos
- Usar `else` para el caso contrario
- Manejar dos alternativas

```python
#Estructura if-else
edad=15
if edad>=18:
    print("eres mayor de edad")
    print("puedes entrar a la disco XD")
else:
    print("eres menor de edad")
    print("no puedes entrar")
    print("Mejor haz tu tarea Oe")
```

```python
# Ejemplo práctico: ¿Número par o impar?

numero=7
if numero%2==0:
    print(f"{numero} es par")
else:
    print(f"{numero} es impar")
```
