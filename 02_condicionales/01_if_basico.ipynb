# 🔀 Estructuras Condicionales - if básico

## 🎯 Objetivos
- Entender qué son las estructuras condicionales
- Aprender a usar la sentencia `if`
- Evaluar condiciones simples

## 📖 ¿Qué son las condicionales?
Las estructuras condicionales nos permiten ejecutar diferentes bloques de código dependiendo de si una condición es verdadera o falsa.

## 🏗️ Sintaxis básica de if

```python
# Ejemplo básico de if
edad = 18

if edad >= 18:
    print("Eres mayor de edad")
    print("Puedes votar")
    print("Puedes conducir")

print("Este mensaje siempre se muestra")
```

```python
# Otro ejemplo
temperatura = 30

if temperatura > 25:
    print("Hace calor ☀️")
    print("Recomiendo ir a la playa")

if temperatura < 15:
    print("Hace frío ❄️")
    print("Abrígate bien")
```

```python
# Importante: la indentación (sangría)
if True:
    print("Este código está dentro del if")
    print("Este también")
print("Este ya está fuera del if")
```

## 💡 Reglas importantes:
1. Siempre termina la línea de `if` con dos puntos `:`
2. El bloque indentado es lo que se ejecuta si la condición es `True`
3. Python es estricto con la indentación (generalmente 4 espacios)
