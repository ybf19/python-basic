### Ejercicio 1: Clasificador de números
Pide un número al usuario y clasifícalo en:
- Positivo / Negativo / Cero
- Par / Impar

```python
# Solución Ejercicio 1
numero = float(input("Ingresa un número: "))

# Clasificación por signo
if numero > 0:
    print("El número es POSITIVO")
elif numero < 0:
    print("El número es NEGATIVO")
else:
    print("El número es CERO")

# Clasificación por paridad (solo para enteros)
if numero % 1 == 0:  # Si es entero
    if numero % 2 == 0:
        print("El número es PAR")
    else:
        print("El número es IMPAR")
```

### Ejercicio 2: Calculadora de Índice de Masa Corporal (IMC)
Pide peso (kg) y altura (m) y calcula el IMC: peso / altura²
Clasifica según:
- < 18.5: Bajo peso
- 18.5 - 24.9: Normal
- 25 - 29.9: Sobrepeso
- >= 30: Obesidad

```python
# Solución Ejercicio 2
peso = float(input("Peso (kg): "))
altura = float(input("Altura (m): "))

imc = peso / (altura ** 2)
print(f"Tu IMC es: {imc:.2f}")

if imc < 18.5:
    print("Clasificación: Bajo peso")
elif imc < 25:
    print("Clasificación: Normal")
elif imc < 30:
    print("Clasificación: Sobrepeso")
else:
    print("Clasificación: Obesidad")
```

### Ejercicio 3: Año bisiesto
Determina si un año es bisiesto:
- Divisible entre 4
- Pero no divisible entre 100, excepto si también es divisible entre 400

```python
# Solución Ejercicio 3
año = int(input("Ingresa un año: "))

if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):
    print(f"{año} es un año BISIESTO ✅")
else:
    print(f"{año} NO es un año bisiesto ❌")
```

### Ejercicio 4: Sistema de calificaciones
Pide una nota del 0 al 100 y muestra:
- 90-100: Excelente (A)
- 80-89: Bueno (B)
- 70-79: Suficiente (C)
- 60-69: Insuficiente (D)
- 0-59: Reprobado (F)
- Fuera de rango: Nota inválida

```python
# Solución Ejercicio 4
nota = float(input("Ingresa tu nota (0-100): "))

if nota < 0 or nota > 100:
    print("❌ Nota inválida")
elif nota >= 90:
    print(f"Excelente (A) - {nota}")
elif nota >= 80:
    print(f"Bueno (B) - {nota}")
elif nota >= 70:
    print(f"Suficiente (C) - {nota}")
elif nota >= 60:
    print(f"Insuficiente (D) - {nota}")
else:
    print(f"Reprobado (F) - {nota}")
```

### Ejercicio 5: Piedra, papel o tijera
Juego contra la computadora (elige aleatorio)
Pide al usuario: piedra, papel o tijera
Determina quién gana

```python
# Solución Ejercicio 5
import random

opciones = ["piedra", "papel", "tijera"]
usuario = input("Elige: piedra, papel o tijera: ").lower()
computadora = random.choice(opciones)

print(f"Computadora eligió: {computadora}")

if usuario not in opciones:
    print("❌ Opción inválida")
elif usuario == computadora:
    print("🤝 Empate!")
elif (usuario == "piedra" and computadora == "tijera") or \
     (usuario == "papel" and computadora == "piedra") or \
     (usuario == "tijera" and computadora == "papel"):
    print("🎉 ¡GANASTE!")
else:
    print("💻 Perdiste... la computadora gana")
```
