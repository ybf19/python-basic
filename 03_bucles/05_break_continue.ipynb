# 🛑 break y continue

## 🎯 Objetivos
- Interrumpir bucles con `break`
- Saltar iteraciones con `continue`
- Control fino sobre la ejecución

```python
# break - sale del bucle completamente
print("break - buscar primer número par:")
numeros = [3, 7, 8, 2, 9, 4]

for num in numeros:
    print(f"  Analizando: {num}")
    if num % 2 == 0:
        print(f"  ✅ Encontré un par: {num}")
        break
print("Bucle terminado")
```

```python
# continue - salta a la siguiente iteración
print("continue - mostrar solo números pares:")
for i in range(1, 11):
    if i % 2 != 0:  # si es impar
        continue    # salta el resto del código
    print(f"  Número par: {i}")
```

```python
# Ejemplo combinado: validación con break
while True:
    respuesta = input("¿Quieres continuar? (s/n): ").lower()
    if respuesta == 'n':
        print("Saliendo...")
        break
    elif respuesta == 's':
        print("Continuamos!")
    else:
        print("Respuesta inválida. Usa 's' o 'n'")
```
