# 🔀 Bucles Anidados

## 🎯 Objetivos
- Poner bucles dentro de otros bucles
- Crear tablas y patrones
- Entender la complejidad de bucles anidados

```python
# Bucles anidados simples
print("Tabla de multiplicar básica:")
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i*j}")
    print("-" * 10)
```

```python
# Matriz 3x3
print("Matriz 3x3:")
for fila in range(3):
    for columna in range(3):
        print(f"({fila},{columna})", end=" ")
    print()  # nueva línea después de cada fila
```

```python
# Reloj digital simple
print("Reloj (primeros 3 segundos):")
for hora in range(24):
    for minuto in range(60):
        for segundo in range(60):
            print(f"\r{hora:02d}:{minuto:02d}:{segundo:02d}", end="")
            # time.sleep(1)  # descomenta para ver el efecto
        if hora == 0 and minuto == 0 and segundo == 2:  # solo 3 segundos
            break
    break
print("\n⏰ ¡Tiempo!")
```
