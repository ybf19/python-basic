# 🎮 PROYECTO: Juego de Adivinanza Mejorado

## 🎯 Objetivo del proyecto
Crear un juego completo de adivinanza con:
- Diferentes niveles de dificultad
- Sistema de puntuación
- Récords
- Opción de jugar múltiples rondas

```python
# 🎮 JUEGO DE ADIVINANZA - VERSIÓN COMPLETA
import random
import datetime

print("=" * 60)
print("            🎮 JUEGO DE ADIVINANZA 2.0")
print("=" * 60)

# Configuración inicial
puntaje_total = 0
rondas_jugadas = 0
records = []

while True:
    print("\n" + "=" * 60)
    print("📋 MENÚ PRINCIPAL")
    print("=" * 60)
    print("1. 🎯 Jugar")
    print("2. 📊 Ver estadísticas")
    print("3. 🏆 Ver récords")
    print("4. ❌ Salir")
    
    opcion = input("\nElige una opción (1-4): ")
    
    if opcion == "1":
        # Configurar partida
        rondas_jugadas += 1
        print("\n" + "-" * 40)
        print("🎮 NUEVA PARTIDA")
        print("-" * 40)
        
        print("Selecciona dificultad:")
        print("1. Fácil (1-50, 10 intentos)")
        print("2. Medio (1-100, 7 intentos)")
        print("3. Difícil (1-200, 5 intentos)")
        
        dificultad = input("Dificultad (1-3): ")
        
        # Configurar según dificultad
        if dificultad == "1":
            maximo = 50
            intentos_max = 10
            print("🎯 Dificultad: FÁCIL")
        elif dificultad == "2":
            maximo = 100
            intentos_max = 7
            print("🎯 Dificultad: MEDIA")
        else:
            maximo = 200
            intentos_max = 5
            print("🎯 Dificultad: DIFÍCIL")
        
        # Iniciar juego
        secreto = random.randint(1, maximo)
        intentos = 0
        adivinado = False
        
        print(f"\n🔢 Número secreto entre 1 y {maximo}")
        print(f"💪 Tienes {intentos_max} intentos")
        
        while intentos < intentos_max and not adivinado:
            intentos_restantes = intentos_max - intentos
            print(f"\n📊 Intentos restantes: {intentos_restantes}")
            
            try:
                intento = int(input(f"Intento #{intentos + 1}: "))
                
                print(f"tu número {intento}: ")
                if intento < 1 or intento > maximo:
                    print(f"⚠️ El número debe estar entre 1 y {maximo}")
                    continue

                intentos += 1
                
                if intento < secreto:
                    print("El número a adivinar es 📈 Mayor")
                elif intento > secreto:
                    print("El número a adivinar es 📉 Menor")
                else:
                    print(f"\n🎉 ¡FELICIDADES! Adivinaste en {intentos} intentos")
                    
                    # Calcular puntaje
                    puntaje = (intentos_max - intentos + 1) * 100
                    puntaje_total += puntaje
                    print(f"🏆 Puntaje de esta ronda: {puntaje}")
                    print(f"📈 Puntaje total: {puntaje_total}")
                    
                    # Guardar récord
                    fecha = datetime.datetime.now().strftime("%d/%m/%Y %H:%M")
                    records.append({
                        "fecha": fecha,
                        "intentos": intentos,
                        "maximo": maximo,
                        "puntaje": puntaje
                    })
                    adivinado = True
                    
            except ValueError:
                print("❌ Ingresa un número válido")
        
        if not adivinado:
            print(f"\n😢 ¡Agotaste los intentos! El número era: {secreto}")
            print("💪 ¡Mejor suerte en la próxima!")
    
    elif opcion == "2":
        print("\n" + "=" * 60)
        print("📊 ESTADÍSTICAS")
        print("=" * 60)
        print(f"🎮 Partidas jugadas: {rondas_jugadas}")
        print(f"🏆 Puntaje total: {puntaje_total}")
        if rondas_jugadas > 0:
            print(f"📈 Promedio por partida: {puntaje_total // rondas_jugadas}")
        print(f"📅 Fecha: {datetime.datetime.now().strftime('%d/%m/%Y %H:%M')}")
    
    elif opcion == "3":
        print("\n" + "=" * 60)
        print("🏆 RÉCORDS")
        print("=" * 60)
        
        if records:
            # Ordenar por puntaje
            records_ordenados = sorted(records, key=lambda x: x["puntaje"], reverse=True)
            
            for i, record in enumerate(records_ordenados[:5], 1):
                print(f"\n{i}. 🎯 Fecha: {record['fecha']}")
                print(f"   Intentos: {record['intentos']}")
                print(f"   Rango: 1-{record['maximo']}")
                print(f"   Puntaje: {record['puntaje']}")
        else:
            print("Aún no hay récords. ¡Juega una partida!")
    
    elif opcion == "4":
        print("\n" + "=" * 60)
        print("👋 ¡GRACIAS POR JUGAR!")
        print("=" * 60)
        print(f"Partidas jugadas hoy: {rondas_jugadas}")
        print(f"Puntaje total: {puntaje_total}")
        print("\n✨ ¡Hasta la próxima!")
        break
    
    else:
        print("❌ Opción inválida. Elige 1-4")

print("\n" + "=" * 60)
```
