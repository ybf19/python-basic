### Ejercicio 1: Suma de números
Pide un número n y suma todos los números del 1 al n.

```python
# Solución Ejercicio 1
n = int(input("Ingresa un número: "))
suma = 0

for i in range(1, n + 1):
    suma += i

print(f"La suma del 1 al {n} es: {suma}")

# Versión con fórmula matemática
suma_formula = n * (n + 1) // 2
print(f"Verificación (fórmula): {suma_formula}")
```

### Ejercicio 2: Factorial
Calcula el factorial de un número (n!)
Ejemplo: 5! = 5 * 4 * 3 * 2 * 1 = 120

```python
# Solución Ejercicio 2
n = int(input("Ingresa un número para calcular factorial: "))
factorial = 1

for i in range(1, n + 1):
    factorial *= i

print(f"{n}! = {factorial}")
```

### Ejercicio 3: Números primos
Determina si un número es primo (solo divisible por 1 y sí mismo)

```python
# Solución Ejercicio 3
n = int(input("Ingresa un número: "))

if n < 2:
    print(f"{n} NO es primo")
else:
    es_primo = True
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            es_primo = False
            break
    
    if es_primo:
        print(f"{n} ES primo ✅")
    else:
        print(f"{n} NO es primo ❌")
```

### Ejercicio 4: Fibonacci
Genera los primeros n números de la secuencia Fibonacci:
0, 1, 1, 2, 3, 5, 8, 13, ...

```python
# Solución Ejercicio 4
n = int(input("¿Cuántos números Fibonacci quieres? "))

a, b = 0, 1
print("Secuencia Fibonacci:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
print()
```

### Ejercicio 5: Adivina el número
La computadora elige un número al azar del 1 al 100.
El usuario tiene que adivinarlo con pistas de "mayor" o "menor".
Cuenta los intentos.

```python
# Solución Ejercicio 5
import random

secreto = random.randint(1, 100)
intentos = 0
adivinado = False

print("🎯 Adivina el número (1-100)")

while not adivinado:
    intento = int(input("Tu intento: "))
    intentos += 1
    
    if intento < secreto:
        print("📈 Mayor")
    elif intento > secreto:
        print("📉 Menor")
    else:
        print(f"🎉 ¡Correcto! Lo adivinaste en {intentos} intentos")
        adivinado = True
```
