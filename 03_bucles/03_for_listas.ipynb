# 📋 for con Listas y Colecciones

## 🎯 Objetivos
- Recorrer listas con for
- Acceder a índices con enumerate()
- Trabajar con múltiples listas

```python
# Recorriendo una lista simple
frutas = ["manzana", "pera", "uva", "naranja"]

print("Lista de frutas:")
for fruta in frutas:
    print(f"  🍎 {fruta}")
```

```python
# Accediendo al índice y valor con enumerate
colores = ["rojo", "verde", "azul", "amarillo"]

print("Colores con índice:")
for indice, color in enumerate(colores):
    print(f"  {indice}: {color}")
```

```python
# Procesando listas numéricas
numeros = [10, 20, 30, 40, 50]
cuadrados = []

print("Números originales:", numeros)
for num in numeros:
    cuadrado = num ** 2
    cuadrados.append(cuadrado)
    print(f"  {num}² = {cuadrado}")

print("Cuadrados:", cuadrados)
```

```python
# Filtrando elementos
edades = [15, 22, 18, 34, 12, 45, 17, 30]

print("Edades:", edades)
print("Mayores de edad:")
for edad in edades:
    if edad >= 18:
        print(f"  ✅ {edad} años")
    else:
        print(f"  ❌ {edad} años (menor)")
```
