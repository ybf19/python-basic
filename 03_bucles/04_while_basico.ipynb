# 🔁 Bucle while

## 🎯 Objetivos
- Entender el bucle `while`
- Crear bucles basados en condiciones
- Evitar bucles infinitos

```python
# while básico
contador = 0

while contador < 5:
    print(f"Contador: {contador}")
    contador += 1  # ¡IMPORTANTE! actualizar la variable

print("¡Bucle terminado!")
```

```python
# while con input del usuario
respuesta = ""

while respuesta != "salir":
    respuesta = input("Escribe algo ('salir' para terminar): ")
    print(f"  Escribiste: {respuesta}")

print("👋 ¡Hasta luego!")
```

```python
# Validación de entrada
edad = -1

while edad < 0 or edad > 120:
    edad = int(input("Ingresa tu edad (0-120): "))
    if edad < 0 or edad > 120:
        print("❌ Edad inválida. Intenta de nuevo.")

print(f"✅ Edad válida: {edad} años")
```

```python
# ¡Cuidado con los bucles infinitos!
# Esto causaría un bucle infinito:
# while True:
#     print("Esto no termina nunca...")

# Forma correcta de hacer un bucle "infinito" controlado
print("Contador regresivo (presiona Ctrl+C para detener):")
contador = 10
while contador > 0:
    print(f"  {contador}")
    contador -= 1
    # import time; time.sleep(1)  # esperar 1 segundo
print("¡Tiempo!")
```
