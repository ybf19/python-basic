# 🔄 Introducción a Bucles - for

## 🎯 Objetivos
- Entender qué son los bucles
- Aprender a usar `for` con strings
- Iterar sobre secuencias simples

## 📖 ¿Qué son los bucles?
Los bucles nos permiten repetir un bloque de código múltiples veces. Son útiles para recorrer secuencias, procesar listas de datos, o repetir acciones.

## 🏗️ Bucle for con strings

```python
# Recorriendo un string letra por letra
nombre = "Python"

print("Recorriendo el nombre:")
for letra in nombre:
    print(f"Letra: {letra}")
```

```python
# Ejemplo más visual
palabra = "HOLA"

print("Desglose de la palabra:")
for caracter in palabra:
    print(f"🔤 {caracter}")
    
print("¡Terminé de recorrer!")
```

```python
# Podemos hacer algo con cada elemento
texto = "Python"

for letra in texto:
    print(f"La letra '{letra}' en mayúscula es '{letra.upper()}'")
```
