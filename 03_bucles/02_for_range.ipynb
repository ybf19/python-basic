# 🔢 for con range()

## 🎯 Objetivos
- Usar la función `range()` para generar secuencias numéricas
- Controlar el inicio, fin y paso de los rangos
- Repetir acciones un número específico de veces

```python
# range(stop) - desde 0 hasta stop-1
print("range(5):")
for i in range(5):
    print(f"  Valor: {i}")

print("\nrange(3):")
for numero in range(3):
    print(f"  Número: {numero}")
```

```python
# range(start, stop) - desde start hasta stop-1
print("range(2, 6):")
for i in range(2, 6):
    print(f"  Valor: {i}")

print("\nrange(10, 16):")
for num in range(10, 16):
    print(f"  Número: {num}")
```

```python
# range(start, stop, step) - con paso personalizado
print("range(0, 10, 2) - números pares:")
for i in range(0, 10, 2):
    print(f"  {i}")

print("\nrange(5, 0, -1) - cuenta regresiva:")
for i in range(5, 0, -1):
    print(f"  {i}")
print("  🚀 ¡Despegue!")
```

```python
# Ejemplos prácticos
print("Tabla del 5:")
for i in range(1, 11):
    print(f"  5 x {i} = {5 * i}")
```

```python
# Sumar números del 1 al 10
suma = 0
for i in range(1, 11):
    suma += i
    print(f"  Sumando {i} → suma parcial: {suma}")
print(f"\n✅ Suma total: {suma}")
```
