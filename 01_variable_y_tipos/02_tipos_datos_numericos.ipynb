# 🔢 Tipos de Datos Numéricos

## 🎯 Objetivos
- Conocer los tipos numéricos en Python
- Operaciones básicas con números
- Diferencias entre int y float

## 📊 Tipos numéricos básicos
- **int**: números enteros (1, 42, -10)
- **float**: números decimales (3.14, -0.5, 2.0)

```python
#Números enteros(int)
edad=30
año=2024
poblacion=8_000_000_000  #se pueden usar guiones bajos para leer mejor los números muy grandes

print(f"Edad: {edad} - Tipo: {type(edad)}")
print(f"Año: {año} - Tipo: {type(año)}")
print(f"Población: {poblacion} - Tipo: {type(poblacion)}")
```

```python
#Números decimales
precio=99.99
pi=3.1415
temperatura=-5.5

print(f"precion {precio} -tipo: {type(precio)}")
print(f"pi {pi} -tipo {type(pi)}")
print(f"{temperatura} -tipo {type(temperatura)}")
```

```python
# Operaciones matemáticas básicas
a = 10
b = 3

print(f"Suma: {a} + {b} = {a + b}")
print(f"Resta: {a} - {b} = {a - b}")
print(f"Multiplicación: {a} * {b} = {a * b}")
print(f"División: {a} / {b} = {a / b}")
print(f"División entera: {a} // {b} = {a // b}")
print(f"Módulo (resto): {a} % {b} = {a % b}")
print(f"Potencia: {a} ** {b} = {a ** b}")
```
