# 📝 Strings (Cadenas de Texto)

## 🎯 Objetivos
- Crear y manipular texto
- Conocer métodos útiles para strings
- Formateo de cadenas

```python
#Formas de crear strings
nombre="ana"
apellido="García"
parrafo="""Este es un texto 
que ocupa varías/múltiples
líneas"""

print(f"Nombre: {nombre}")
print(f"Apellido: {apellido}")
print(f"Parrafo: \n{parrafo}")
```

```python
# Concatenación y repetición
saludo = "Hola" + " " + "Mundo"
print(saludo)

eco = "Ja" * 3 + "!"
print(eco)
```

```python
# Métodos útiles de strings
texto = "  Python es GENIAL  "

print(f"Original: '{texto}'")
print(f"minusculas: '{texto.lower()}'")
print(f"MAYUSCULAS: '{texto.upper()}'")
print(f"Sin espacios: '{texto.strip()}'")
print(f"Reemplazar: '{texto.replace('GENIAL', 'increíble')}'")
print(f"Largo: {len(texto)} caracteres")
```

```python
# Formateo de strings (f-strings)
nombre = "Ana"
edad = 25

# f-strings (recomendado)
print(f"Me llamo {nombre} y tengo {edad} años")

# .format()
print("Me llamo {} y tengo {} años".format(nombre, edad))

# % formateo (antiguo)
print("Me llamo %s y tengo %d años" % (nombre, edad))
```
