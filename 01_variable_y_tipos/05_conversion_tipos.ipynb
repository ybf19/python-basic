# 🔄 Conversión de Tipos (Type Casting)

## 🎯 Objetivos
- Convertir entre diferentes tipos de datos
- Entender cuándo es necesario convertir
- Evitar errores comunes

```python
# Conversión a entero (int)
print(int(3.14))        # float → int: 3
print(int("42"))         # string → int: 42
print(int(True))         # bool → int: 1
print(int(False))        # bool → int: 0
```

```python
# Conversión a float
print(float(5))          # int → float: 5.0
print(float("3.14"))     # string → float: 3.14
print(float("42"))       # string → float: 42.0
```

```python
# Conversión a string
print(str(42))           # int → string: "42"
print(str(3.14))         # float → string: "3.14"
print(str(True))         # bool → string: "True"
```

```python
# Conversión a booleano
print(bool(0))           # False
print(bool(1))           # True
print(bool(42))          # True (cualquier número ≠ 0)
print(bool(""))          # False (string vacío)
print(bool("Hola"))      # True (string no vacío)
print(bool(None))        # False
```

```python
# Ejemplo práctico: input siempre devuelve string
edad_input = input("¿Cuántos años tienes? ")
print(f"Tipo: {type(edad_input)}")

# Convertimos para poder hacer operaciones matemáticas
edad = int(edad_input)
print(f"El año que viene tendrás {edad + 1} años")
```
