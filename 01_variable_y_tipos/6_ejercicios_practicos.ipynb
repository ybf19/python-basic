## Ejercicio 2: Calculadora de propina
Pide al usuario el total de una cuenta y calcula:
- Propina del 10%
- Propina del 15%
- Propina del 20%
Muestra todos los resultados.

```python
# Escribe tu solución aquí
total = float(input("Total de la cuenta: $"))#ejemplo con 25

propina_10 = total * 0.10
propina_15 = total * 0.15
propina_20 = total * 0.20

print(f"Propina 10%: ${propina_10:.2f}")
print(f"Propina 15%: ${propina_15:.2f}")
print(f"Propina 20%: ${propina_20:.2f}")
```

## Ejercicio 3: Conversor de temperaturas
Convierte temperatura de Celsius a Fahrenheit y Kelvin.
Fórmulas:
- Fahrenheit = (Celsius * 9/5) + 32
- Kelvin = Celsius + 273.15

```python
# Escribe tu solución aquí
celsius = float(input("Temperatura en Celsius: "))

fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15

print(f"{celsius}°C = {fahrenheit:.2f}°F")
print(f"{celsius}°C = {kelvin:.2f}K")
```
