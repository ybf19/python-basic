## 📦 ¿Qué es una variable?
Una variable es como una **caja** donde guardamos información. 
Le ponemos una etiqueta (nombre) y dentro guardamos un valor.

## 🏗️ Creando nuestras primeras variables

```python
#creando variables
nombre="ana"
edad=25
altura=1.65
#mostrar contenido
print("nombre", nombre)
print("edad", edad)
print("altura", altura)
```

## 📏 Reglas para nombrar variables
✅ **Válido:** `nombre`, `edad_persona`, `altura2`, `miVariable`
❌ **Inválido:** `2altura` (empieza con número), `mi-variable` (guión), `class` (palabra reservada)

## 💡 Buenas prácticas
- Usar nombres descriptivos: `edad_usuario` en lugar de `e`
- Usar snake_case: `nombre_completo` (recomendado en Python)
