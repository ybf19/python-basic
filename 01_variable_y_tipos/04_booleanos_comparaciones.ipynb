# ✅ Booleanos y Operadores de Comparación

## 🎯 Objetivos
- Entender el tipo bool (True/False)
- Usar operadores de comparación
- Operadores lógicos (and, or, not)

```python
# El tipo booleano
es_mayor = True
tiene_permiso = False

print(f"¿Es mayor? {es_mayor} - Tipo: {type(es_mayor)}")
print(f"¿Tiene permiso? {tiene_permiso} - Tipo: {type(tiene_permiso)}")
```

```python
# Operadores de comparación
edad = 18

print(f"edad = {edad}")
print(f"¿Es igual a 18? {edad == 18}")
print(f"¿Es diferente de 18? {edad != 18}")
print(f"¿Es mayor que 18? {edad > 18}")
print(f"¿Es menor que 18? {edad < 18}")
print(f"¿Es mayor o igual que 18? {edad >= 18}")
print(f"¿Es menor o igual que 18? {edad <= 18}")
```

```python
# Operadores lógicos
tiene_dinero = True
hace_buen_dia = False

# AND: ambas deben ser True
print(f"¿Puede salir? {tiene_dinero and hace_buen_dia}")

# OR: al menos una debe ser True
print(f"¿Tiene alguna excusa? {tiene_dinero or hace_buen_dia}")

# NOT: invierte el valor
print(f"¿No tiene dinero? {not tiene_dinero}")
```
